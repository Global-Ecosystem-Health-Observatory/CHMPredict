```python
import pandas as pd
import requests
import os

urls_df = pd.read_csv('/Users/anisr/Downloads/paituli_72727328.txt', header=None, names=['URL'])
urls_df = urls_df.drop(index=0).reset_index(drop=True)  # Drop first entry
urls_df['Filename'] = urls_df['URL'].apply(lambda x: x.split('/')[-1].replace('.jp2', ''))

details_df = pd.read_csv('/Users/anisr/Downloads/chm_list.txt', delim_whitespace=True, header=None)
details_df['Filename'] = details_df.iloc[:, -1].apply(lambda x: x.replace('CHM_', '').replace('_2016.tif', ''))

matching_filenames = urls_df[urls_df['Filename'].isin(details_df['Filename'])]

save_dir = "/Users/anisr/Documents/CHM_Images/Images/"
os.makedirs(save_dir, exist_ok=True)

for _, row in matching_filenames.iterrows():
    filename = row['Filename']
    file_path = os.path.join(save_dir, filename + '.jp2')

    if os.path.exists(file_path):
        print(f"File already exists, skipping: {filename}")
        continue

    url = row['URL']
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Check if the request was successful
        
        with open(file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        
        print(f"Downloaded: {filename}")
    
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}: {e}")
```
